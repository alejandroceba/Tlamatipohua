````python
anos = np.arange(2003,2022+1)
for ano in anos:
    f = f"../datos/Grado_de_Presion/Grado_de_Presion_{ano}.xlsx"
    if ano == 2003:
        gp = pd.read_excel(f,usecols=["Región Hidrológico-Administrativa","Grado de Presión (%)"])
        gp.columns = ["RHA",ano]
    else:
        gp_ano = pd.read_excel(f,usecols=["Grado de Presión (%)"])
        gp_ano.columns = [ano]
        gp[ano] = gp_ano[ano]
gp["RHA"] = gp["RHA"].str.replace(r"^\s*[IVXLCDM]+\s*[-.]?\s*", "", regex=True)
name_mapping = {
    "Península de Baja California": "Península De Baja California",
    'Cuencas Centrales del Norte': 'Cuencas Centrales Del Norte',
    "Península de Yucatán": "Península De Yucatán",
    'Aguas del Valle de México': 'Valle de México'
}
gp['RHA'] = gp['RHA'].replace(name_mapping)
gp
````

````python
from pathlib import Path
import re
import xml.etree.ElementTree as ET

import pandas as pd
import requests

requests.packages.urllib3.disable_warnings()  # notebook runs against an internal HTTPS endpoint

base_dir = Path("../datos")

source_configs = {
    "disponibilidad_acuiferos": {
        "folder": "disponibilidad_acuiferos",
        "uuid": "c9d72f33-5caf-4641-8620-155da91ba82c",
        "layer": "Sina:m06_acuiferos_disponibilidad",
        "service_name": "m06_acuiferos_disponibilidad",
        "kind": "disponibilidad subterránea",
    },
    "disponiblidad_superficial": {
        "folder": "disponiblidad_superficial",
        "uuid": "9b0354df-90e3-4b8e-a656-31528f970fac",
        "layer": "Sina:m05_cuencas_disponibilidad",
        "service_name": "m05_cuencas_disponibilidad",
        "kind": "disponibilidad superficial",
    },
    "repda_subterraneo": {
        "folder": "repda_subterraneo",
        "uuid": "763c4995-8607-4c74-af0e-112c9c1f5e50",
        "layer": "Sina:m01_usoagrupadosbm",
        "service_name": "m01_usoagrupadosbm",
        "kind": "REPDA subterráneo",
    },
    "repda_superficial": {
        "folder": "repda_superficial",
        "uuid": "e29b028e-d335-4c48-abe5-b2dbe8bcc2dc",
        "layer": "Sina:m01_usoagrupadospm",
        "service_name": "m01_usoagrupadospm",
        "kind": "REPDA superficial",
    },
}

ns = {
    "gmd": "http://www.isotc211.org/2005/gmd",
    "gco": "http://www.isotc211.org/2005/gco",
}


def _first_text(root, xpath):
    node = root.find(xpath, ns)
    if node is None:
        return None
    return (node.text or "").strip() or None


def _all_text(root, xpath):
    values = []
    for node in root.findall(xpath, ns):
        value = (node.text or "").strip()
        if value:
            values.append(value)
    return values


def _read_metadata(metadata_path):
    root = ET.parse(metadata_path).getroot()
    return {
        "title": _first_text(root, ".//gmd:identificationInfo//gmd:citation//gmd:title/gco:CharacterString") or "",
        "abstract": _first_text(root, ".//gmd:identificationInfo//gmd:abstract/gco:CharacterString") or "",
        "purpose": _first_text(root, ".//gmd:identificationInfo//gmd:purpose/gco:CharacterString") or "",
        "dates": _all_text(root, ".//gmd:identificationInfo//gmd:date//gco:Date") or _all_text(root, ".//gmd:identificationInfo//gmd:date//gco:DateTime"),
        "urls": _all_text(root, ".//gmd:CI_OnlineResource//gmd:URL/gmd:URL") or _all_text(root, ".//gmd:CI_OnlineResource//gmd:URL"),
    }


def _infer_properties(cfg):
    service_url = f"https://geosinav30.conagua.gob.mx:8080/geoserver/Sina/{cfg['service_name']}/wfs"
    params = {
        "service": "WFS",
        "version": "2.0.0",
        "request": "GetFeature",
        "typeNames": cfg["layer"],
        "outputFormat": "application/json",
        "count": 1,
        "srsname": "EPSG:4326",
    }
    response = requests.get(service_url, params=params, verify=False, timeout=90)
    response.raise_for_status()
    payload = response.json()
    features = payload.get("features", [])
    if not features:
        return []
    return sorted(features[0].get("properties", {}).keys())


schema_rows = []
for source_name, cfg in source_configs.items():
    source_dir = base_dir / cfg["folder"]
    metadata_path = source_dir / cfg["uuid"] / "metadata" / "metadata.xml"
    metadata = _read_metadata(metadata_path)
    local_files = sorted(
        str(path.relative_to(base_dir))
        for path in source_dir.rglob("*")
        if path.is_file()
    )
    try:
        properties = _infer_properties(cfg)
    except Exception as exc:
        properties = [f"<schema inference failed: {type(exc).__name__}>" ]
    schema_rows.append(
        {
            "source": source_name,
            "kind": cfg["kind"],
            "folder": cfg["folder"],
            "metadata_title": metadata["title"],
            "metadata_dates": ", ".join(metadata["dates"]),
            "available_urls": " | ".join(metadata["urls"][:3]),
            "local_files": ", ".join(local_files),
            "wfs_layer": cfg["layer"],
            "inferred_fields": ", ".join(properties),
            "field_count": len(properties),
        }
    )

schema_inventory = pd.DataFrame(schema_rows)
pd.set_option("display.max_colwidth", 120)
pd.set_option("display.max_columns", None)
schema_inventory
````

````python
geo_patterns = {
    "rha": r"\bclvrha\b|\brha\b",
    "rh": r"\bclvrh\b|\brh\b",
    "cuenca": r"\bcuenca\b|\bclvcuenca\b",
    "estado": r"\bclvedo\b|\bestado\b|\bentidad\b",
    "municipio": r"\bclvmun\b|\bmunicipio\b",
}


def _split_fields(field_string):
    return [field.strip() for field in field_string.split(",") if field.strip()]


def _detect_matches(field_string, patterns):
    fields = _split_fields(field_string)
    hits = []
    for label, pattern in patterns.items():
        if any(re.search(pattern, field, flags=re.IGNORECASE) for field in fields):
            hits.append(label)
    return ", ".join(hits) if hits else "none"


schema_inventory["geo_levels_detected"] = schema_inventory["inferred_fields"].apply(
    lambda field_string: _detect_matches(field_string, geo_patterns)
)
schema_inventory["time_fields_detected"] = schema_inventory["inferred_fields"].apply(
    lambda field_string: ", ".join(
        field
        for field in _split_fields(field_string)
        if re.search(r"\banio\b|\bfechacorte\b|\bfechapub\b", field, flags=re.IGNORECASE)
    ) or "none"
)

schema_inventory[["source", "kind", "field_count", "geo_levels_detected", "time_fields_detected"]]
````

## Step 3. Build geo crosswalks

Build reusable geographic crosswalk tables from source-native keys.


````python
import geopandas as gpd

WFS_BASE = "https://geosinav30.conagua.gob.mx:8080/geoserver/Sina"


def fetch_wfs(layer_name, count=100000):
    response = requests.get(
        f"{WFS_BASE}/{layer_name}/wfs",
        params={
            "service": "WFS",
            "version": "2.0.0",
            "request": "GetFeature",
            "typeNames": f"Sina:{layer_name}",
            "outputFormat": "application/json",
            "srsname": "EPSG:4326",
            "count": count,
        },
        verify=False,
        timeout=120,
    )
    response.raise_for_status()
    payload = response.json()
    return gpd.GeoDataFrame.from_features(payload["features"])


def safe_pick(df, cols):
    existing = [c for c in cols if c in df.columns]
    return df[existing].copy(), existing


# Source pulls for crosswalk construction
repda_sub_raw = fetch_wfs("m01_usoagrupadosbm", count=150000)
repda_sup_raw = fetch_wfs("m01_usoagrupadospm", count=150000)
superficial_raw = fetch_wfs("m05_cuencas_disponibilidad", count=30000)
acuiferos_raw = fetch_wfs("m06_acuiferos_disponibilidad", count=30000)

print("Rows fetched:")
print(
    pd.DataFrame(
        {
            "dataset": [
                "repda_subterraneo",
                "repda_superficial",
                "disponibilidad_superficial",
                "disponibilidad_acuiferos",
            ],
            "rows": [
                len(repda_sub_raw),
                len(repda_sup_raw),
                len(superficial_raw),
                len(acuiferos_raw),
            ],
        }
    )
)
````

````python
def extract_year(series):
    return pd.to_numeric(
        series.astype(str).str.extract(r"(19\d{2}|20\d{2})")[0],
        errors="coerce",
    ).astype("Int64")


# 1) Municipio -> Estado -> RHA crosswalk (from REPDA sources)
repda_keys = ["clvmun", "municipio", "clvedo", "estado", "clvrha", "rha", "corterepda"]
repda_sub_keys, repda_sub_existing = safe_pick(repda_sub_raw, repda_keys)
repda_sup_keys, repda_sup_existing = safe_pick(repda_sup_raw, repda_keys)

repda_crosswalk_long = pd.concat(
    [repda_sub_keys.assign(source="repda_subterraneo"), repda_sup_keys.assign(source="repda_superficial")],
    ignore_index=True,
)
if "corterepda" in repda_crosswalk_long.columns:
    repda_crosswalk_long["anio"] = extract_year(repda_crosswalk_long["corterepda"])
else:
    repda_crosswalk_long["anio"] = pd.Series(pd.NA, dtype="Int64")

muni_state_rha_crosswalk = (
    repda_crosswalk_long
    .dropna(subset=[c for c in ["clvmun", "clvedo"] if c in repda_crosswalk_long.columns], how="all")
    .drop_duplicates()
    .sort_values([c for c in ["anio", "clvmun", "clvedo"] if c in repda_crosswalk_long.columns])
    .reset_index(drop=True)
)

# 2) Cuenca -> RH -> RHA crosswalk (from disponibilidad superficial)
surface_keys = ["clvcuenca", "cuenca", "clvrh", "rh", "clvrha", "rha", "anio"]
surface_cx, surface_existing = safe_pick(superficial_raw, surface_keys)

cuenca_rh_rha_crosswalk = (
    surface_cx
    .dropna(subset=[c for c in ["clvcuenca", "clvrh", "clvrha"] if c in surface_cx.columns], how="all")
    .drop_duplicates()
    .sort_values([c for c in ["anio", "clvcuenca", "clvrh", "clvrha"] if c in surface_cx.columns])
    .reset_index(drop=True)
)

# 3) Acuífero -> RHA bridge (from disponibilidad acuíferos)
aquifer_keys = ["clvacuifero", "acuifero", "clvrha", "rha", "anio"]
aquifer_cx, aquifer_existing = safe_pick(acuiferos_raw, aquifer_keys)

acuifero_rha_crosswalk = (
    aquifer_cx
    .dropna(subset=[c for c in ["clvacuifero", "clvrha"] if c in aquifer_cx.columns], how="all")
    .drop_duplicates()
    .sort_values([c for c in ["anio", "clvacuifero", "clvrha"] if c in aquifer_cx.columns])
    .reset_index(drop=True)
)

crosswalk_inventory = pd.DataFrame(
    [
        {
            "crosswalk": "municipio_estado_rha",
            "rows": len(muni_state_rha_crosswalk),
            "columns": ", ".join(muni_state_rha_crosswalk.columns),
            "sources": ", ".join(sorted(repda_crosswalk_long["source"].dropna().unique().tolist())),
        },
        {
            "crosswalk": "cuenca_rh_rha",
            "rows": len(cuenca_rh_rha_crosswalk),
            "columns": ", ".join(cuenca_rh_rha_crosswalk.columns),
            "sources": "disponibilidad_superficial",
        },
        {
            "crosswalk": "acuifero_rha",
            "rows": len(acuifero_rha_crosswalk),
            "columns": ", ".join(acuifero_rha_crosswalk.columns),
            "sources": "disponibilidad_acuiferos",
        },
    ]
)

print("Crosswalk inventory")
display(crosswalk_inventory)

print("\nSample: municipio_estado_rha")
display(muni_state_rha_crosswalk.head(10))

print("\nSample: cuenca_rh_rha")
display(cuenca_rh_rha_crosswalk.head(10))

print("\nSample: acuifero_rha")
display(acuifero_rha_crosswalk.head(10))
````

## Step 4. Prepare component tables

Create normalized component tables for extraction and availability terms before indicator calculation.


````python
def to_numeric_safe(df, cols):
    out = df.copy()
    for col in cols:
        if col in out.columns:
            out[col] = pd.to_numeric(out[col], errors="coerce")
    return out


# TFWW candidate components from REPDA
repda_common_cols = ["clvmun", "municipio", "clvedo", "estado", "clvrha", "rha", "corterepda", "total"]
repda_sub_tfww = to_numeric_safe(repda_sub_raw[repda_common_cols].copy(), ["total"]).assign(
    anio=extract_year(repda_sub_raw["corterepda"]),
    component="csub_repda_total_hm3",
    geo_base="municipio",
    value=lambda d: d["total"],
)
repda_sup_tfww = to_numeric_safe(repda_sup_raw[repda_common_cols].copy(), ["total"]).assign(
    anio=extract_year(repda_sup_raw["corterepda"]),
    component="csurf_repda_total_hm3",
    geo_base="municipio",
    value=lambda d: d["total"],
)

tfww_components = pd.concat([repda_sub_tfww, repda_sup_tfww], ignore_index=True)
tfww_components = tfww_components[
    ["anio", "component", "geo_base", "clvmun", "municipio", "clvedo", "estado", "clvrha", "rha", "value"]
].dropna(subset=["anio", "value"])


# Surface availability and related terms (cuenca base)
surface_component_cols = [
    "anio", "clvcuenca", "cuenca", "clvrh", "rh", "clvrha", "rha",
    "cp", "ar", "im", "ex", "ab", "ev", "av", "r", "rxy", "d",
]
surface_available = [c for c in surface_component_cols if c in superficial_raw.columns]
surface_terms = to_numeric_safe(superficial_raw[surface_available].copy(), ["cp", "ar", "im", "ex", "ab", "ev", "av", "r", "rxy", "d"])

surface_long = surface_terms.melt(
    id_vars=[c for c in ["anio", "clvcuenca", "cuenca", "clvrh", "rh", "clvrha", "rha"] if c in surface_terms.columns],
    value_vars=[c for c in ["cp", "ar", "im", "ex", "ab", "ev", "av", "r", "rxy", "d"] if c in surface_terms.columns],
    var_name="component",
    value_name="value",
).assign(geo_base="cuenca")

surface_long["component"] = "superficial_" + surface_long["component"].astype(str)
surface_long = surface_long.dropna(subset=["anio", "value"])


# Aquifer availability and environmental terms (acuifero base)
aquifer_component_cols = ["anio", "clvacuifero", "acuifero", "clvrha", "rha", "r", "dnc", "dmapositiva", "dmanegativa"]
aquifer_available = [c for c in aquifer_component_cols if c in acuiferos_raw.columns]
aquifer_terms = to_numeric_safe(acuiferos_raw[aquifer_available].copy(), ["r", "dnc", "dmapositiva", "dmanegativa"])

aquifer_long = aquifer_terms.melt(
    id_vars=[c for c in ["anio", "clvacuifero", "acuifero", "clvrha", "rha"] if c in aquifer_terms.columns],
    value_vars=[c for c in ["r", "dnc", "dmapositiva", "dmanegativa"] if c in aquifer_terms.columns],
    var_name="component",
    value_name="value",
).assign(geo_base="acuifero")

aquifer_long["component"] = "acuifero_" + aquifer_long["component"].astype(str)
aquifer_long = aquifer_long.dropna(subset=["anio", "value"])


component_inventory = pd.DataFrame(
    [
        {
            "table": "tfww_components",
            "rows": len(tfww_components),
            "components": ", ".join(sorted(tfww_components["component"].unique().tolist())),
            "years_min": int(tfww_components["anio"].min()) if len(tfww_components) else None,
            "years_max": int(tfww_components["anio"].max()) if len(tfww_components) else None,
        },
        {
            "table": "surface_long",
            "rows": len(surface_long),
            "components": ", ".join(sorted(surface_long["component"].unique().tolist())),
            "years_min": int(surface_long["anio"].min()) if len(surface_long) else None,
            "years_max": int(surface_long["anio"].max()) if len(surface_long) else None,
        },
        {
            "table": "aquifer_long",
            "rows": len(aquifer_long),
            "components": ", ".join(sorted(aquifer_long["component"].unique().tolist())),
            "years_min": int(aquifer_long["anio"].min()) if len(aquifer_long) else None,
            "years_max": int(aquifer_long["anio"].max()) if len(aquifer_long) else None,
        },
    ]
)

display(component_inventory)
````

## Step 5. Calculate indicator series (RHA proxy)

Proxy operationalization for now:

- TFWW = csurf_repda_total_hm3 + csub_repda_total_hm3
- TRWR_proxy = superficial_d + acuifero_dmapositiva
- EFR_proxy = superficial_rxy + acuifero_dnc
- estres_proxy = TFWW / (TRWR_proxy - EFR_proxy) * 100


````python
import numpy as np

# Build a robust RHA label in each component family

tfww_rha_base = tfww_components.copy()
tfww_rha_base["rha_name"] = tfww_rha_base["rha"] if "rha" in tfww_rha_base.columns else tfww_rha_base["clvrha"]
tfww_rha_base["rha_name"] = tfww_rha_base["rha_name"].fillna(tfww_rha_base["clvrha"])

tfww_rha = (
    tfww_rha_base
    .dropna(subset=["anio", "clvrha", "value"])
    .groupby(["anio", "clvrha", "component"], as_index=False)["value"]
    .sum()
    .pivot_table(index=["anio", "clvrha"], columns="component", values="value", fill_value=0)
    .reset_index()
)

surface_rha_base = surface_long.copy()
surface_rha_base["rha_name"] = surface_rha_base["rha"] if "rha" in surface_rha_base.columns else surface_rha_base["clvrha"]
surface_rha_base["rha_name"] = surface_rha_base["rha_name"].fillna(surface_rha_base["clvrha"])

surface_rha = (
    surface_rha_base[surface_rha_base["component"].isin(["superficial_d", "superficial_rxy"])]
    .dropna(subset=["anio", "clvrha", "value"])
    .groupby(["anio", "clvrha", "component"], as_index=False)["value"]
    .sum()
    .pivot_table(index=["anio", "clvrha"], columns="component", values="value", fill_value=0)
    .reset_index()
)

aquifer_rha_base = aquifer_long.copy()
aquifer_rha_base["rha_name"] = aquifer_rha_base["rha"] if "rha" in aquifer_rha_base.columns else aquifer_rha_base["clvrha"]
aquifer_rha_base["rha_name"] = aquifer_rha_base["rha_name"].fillna(aquifer_rha_base["clvrha"])

aquifer_rha = (
    aquifer_rha_base[aquifer_rha_base["component"].isin(["acuifero_dmapositiva", "acuifero_dnc"])]
    .dropna(subset=["anio", "clvrha", "value"])
    .groupby(["anio", "clvrha", "component"], as_index=False)["value"]
    .sum()
    .pivot_table(index=["anio", "clvrha"], columns="component", values="value", fill_value=0)
    .reset_index()
)

rha_proxy_panel = (
    tfww_rha
    .merge(surface_rha, on=["anio", "clvrha"], how="inner")
    .merge(aquifer_rha, on=["anio", "clvrha"], how="inner")
)

# Attach a canonical RHA name from available crosswalk tables
rha_names = pd.concat(
    [
        tfww_rha_base[["clvrha", "rha_name"]],
        surface_rha_base[["clvrha", "rha_name"]],
        aquifer_rha_base[["clvrha", "rha_name"]],
    ],
    ignore_index=True,
).dropna().drop_duplicates(subset=["clvrha", "rha_name"])

rha_name_map = (
    rha_names.groupby("clvrha")["rha_name"]
    .agg(lambda s: s.mode().iloc[0] if not s.mode().empty else s.iloc[0])
    .to_dict()
)
rha_proxy_panel["rha_name"] = rha_proxy_panel["clvrha"].map(rha_name_map).fillna(rha_proxy_panel["clvrha"])

required_components = [
    "csurf_repda_total_hm3",
    "csub_repda_total_hm3",
    "superficial_d",
    "superficial_rxy",
    "acuifero_dmapositiva",
    "acuifero_dnc",
]

for col in required_components:
    if col not in rha_proxy_panel.columns:
        rha_proxy_panel[col] = 0.0

rha_proxy_panel["TFWW_hm3"] = (
    rha_proxy_panel["csurf_repda_total_hm3"] + rha_proxy_panel["csub_repda_total_hm3"]
)
rha_proxy_panel["TRWR_proxy_hm3"] = (
    rha_proxy_panel["superficial_d"] + rha_proxy_panel["acuifero_dmapositiva"]
)
rha_proxy_panel["EFR_proxy_hm3"] = (
    rha_proxy_panel["superficial_rxy"] + rha_proxy_panel["acuifero_dnc"]
)
rha_proxy_panel["available_proxy_hm3"] = (
    rha_proxy_panel["TRWR_proxy_hm3"] - rha_proxy_panel["EFR_proxy_hm3"]
)

rha_proxy_panel["estres_proxy_pct"] = np.where(
    rha_proxy_panel["available_proxy_hm3"] > 0,
    rha_proxy_panel["TFWW_hm3"] / rha_proxy_panel["available_proxy_hm3"] * 100,
    np.nan,
)

rha_proxy_panel = rha_proxy_panel.sort_values(["anio", "clvrha"]).reset_index(drop=True)

rha_indicator_preview = rha_proxy_panel[
    [
        "anio", "clvrha", "rha_name", "TFWW_hm3", "TRWR_proxy_hm3",
        "EFR_proxy_hm3", "available_proxy_hm3", "estres_proxy_pct",
    ]
]

display(rha_indicator_preview.head(20))
````

````python
## Step 6. Validate overlaps and gaps

year_coverage = pd.DataFrame(
    {
        "table": ["tfww_components", "surface_long", "aquifer_long", "rha_proxy_panel"],
        "year_min": [
            int(tfww_components["anio"].min()),
            int(surface_long["anio"].min()),
            int(aquifer_long["anio"].min()),
            int(rha_proxy_panel["anio"].min()) if len(rha_proxy_panel) else None,
        ],
        "year_max": [
            int(tfww_components["anio"].max()),
            int(surface_long["anio"].max()),
            int(aquifer_long["anio"].max()),
            int(rha_proxy_panel["anio"].max()) if len(rha_proxy_panel) else None,
        ],
        "rows": [len(tfww_components), len(surface_long), len(aquifer_long), len(rha_proxy_panel)],
    }
)

overlap_years = sorted(
    set(tfww_components["anio"].dropna().astype(int).unique())
    & set(surface_long["anio"].dropna().astype(int).unique())
    & set(aquifer_long["anio"].dropna().astype(int).unique())
)

rha_gap_check = (
    rha_proxy_panel.assign(
        missing_required=lambda d: d[
            [
                "TFWW_hm3",
                "TRWR_proxy_hm3",
                "EFR_proxy_hm3",
                "available_proxy_hm3",
                "estres_proxy_pct",
            ]
        ].isna().any(axis=1)
    )
    .groupby("anio", as_index=False)["missing_required"]
    .sum()
    .rename(columns={"missing_required": "rows_with_missing_values"})
)

print("Year coverage")
display(year_coverage)
print(f"Overlap years across all three component families: {overlap_years}")
print("\nRHA panel missingness by year")
display(rha_gap_check)
````

## QC. Indicator range and component sanity checks

Check whether computed stress values fall in expected percent bounds and test a stricter proxy specification.


````python
# 1) Current proxy range
current_stats = pd.DataFrame(
    {
        "metric": [
            "rows_total",
            "rows_non_null_indicator",
            "min_indicator",
            "max_indicator",
            "rows_above_100",
            "rows_below_0",
        ],
        "value": [
            len(rha_proxy_panel),
            rha_proxy_panel["estres_proxy_pct"].notna().sum(),
            rha_proxy_panel["estres_proxy_pct"].min(skipna=True),
            rha_proxy_panel["estres_proxy_pct"].max(skipna=True),
            (rha_proxy_panel["estres_proxy_pct"] > 100).sum(skipna=True),
            (rha_proxy_panel["estres_proxy_pct"] < 0).sum(skipna=True),
        ],
    }
)

# 2) Alternative components
# TFWW_consuntivo = total - ecologico - hidroelectrico (when those cols exist)
non_consumptive_labels = [
    "Ecológico",
    "Hidroeléctricas",
]


def get_non_consumptive_cols(df):
    cols = []
    for c in df.columns:
        c_norm = str(c).lower()
        if "ecol" in c_norm or "hidro" in c_norm:
            cols.append(c)
    return cols


sub_non_cons_cols = get_non_consumptive_cols(repda_sub_raw)
sup_non_cons_cols = get_non_consumptive_cols(repda_sup_raw)

repda_sub_alt = repda_sub_raw.copy()
repda_sup_alt = repda_sup_raw.copy()

for c in ["total"] + sub_non_cons_cols:
    if c in repda_sub_alt.columns:
        repda_sub_alt[c] = pd.to_numeric(repda_sub_alt[c], errors="coerce")
for c in ["total"] + sup_non_cons_cols:
    if c in repda_sup_alt.columns:
        repda_sup_alt[c] = pd.to_numeric(repda_sup_alt[c], errors="coerce")

repda_sub_alt["anio"] = extract_year(repda_sub_alt["corterepda"])
repda_sup_alt["anio"] = extract_year(repda_sup_alt["corterepda"])

repda_sub_alt["consuntivo"] = repda_sub_alt["total"] - repda_sub_alt[sub_non_cons_cols].sum(axis=1, min_count=1)
repda_sup_alt["consuntivo"] = repda_sup_alt["total"] - repda_sup_alt[sup_non_cons_cols].sum(axis=1, min_count=1)

# TRWR_alt = superficial_cp + superficial_ar + superficial_im - superficial_ex + acuifero_r
# EFR_alt  = superficial_rxy + acuifero_dnc
surface_alt = superficial_raw.copy()
for c in ["cp", "ar", "im", "ex", "rxy"]:
    if c in surface_alt.columns:
        surface_alt[c] = pd.to_numeric(surface_alt[c], errors="coerce")

surface_alt["trwr_surface_alt"] = (
    surface_alt.get("cp", 0) + surface_alt.get("ar", 0) + surface_alt.get("im", 0) - surface_alt.get("ex", 0)
)
surface_alt["efr_surface_alt"] = surface_alt.get("rxy", 0)

aquifer_alt = acuiferos_raw.copy()
for c in ["r", "dnc"]:
    if c in aquifer_alt.columns:
        aquifer_alt[c] = pd.to_numeric(aquifer_alt[c], errors="coerce")

# Aggregate at RHA-year
alt_tfww = (
    pd.concat(
        [
            repda_sub_alt[["anio", "clvrha", "consuntivo"]],
            repda_sup_alt[["anio", "clvrha", "consuntivo"]],
        ],
        ignore_index=True,
    )
    .dropna(subset=["anio", "clvrha", "consuntivo"])
    .groupby(["anio", "clvrha"], as_index=False)["consuntivo"]
    .sum()
    .rename(columns={"consuntivo": "TFWW_consuntivo_hm3"})
)

alt_surface = (
    surface_alt[["anio", "clvrha", "trwr_surface_alt", "efr_surface_alt"]]
    .dropna(subset=["anio", "clvrha"])
    .groupby(["anio", "clvrha"], as_index=False)[["trwr_surface_alt", "efr_surface_alt"]]
    .sum()
)

alt_aquifer = (
    aquifer_alt[["anio", "clvrha", "r", "dnc"]]
    .dropna(subset=["anio", "clvrha"])
    .groupby(["anio", "clvrha"], as_index=False)[["r", "dnc"]]
    .sum()
    .rename(columns={"r": "trwr_aquifer_alt", "dnc": "efr_aquifer_alt"})
)

rha_alt = (
    alt_tfww
    .merge(alt_surface, on=["anio", "clvrha"], how="inner")
    .merge(alt_aquifer, on=["anio", "clvrha"], how="inner")
)

rha_alt["TRWR_alt_hm3"] = rha_alt["trwr_surface_alt"] + rha_alt["trwr_aquifer_alt"]
rha_alt["EFR_alt_hm3"] = rha_alt["efr_surface_alt"] + rha_alt["efr_aquifer_alt"]
rha_alt["available_alt_hm3"] = rha_alt["TRWR_alt_hm3"] - rha_alt["EFR_alt_hm3"]
rha_alt["estres_alt_pct_raw"] = np.where(
    rha_alt["available_alt_hm3"] > 0,
    rha_alt["TFWW_consuntivo_hm3"] / rha_alt["available_alt_hm3"] * 100,
    np.nan,
)
rha_alt["estres_alt_pct_clipped"] = rha_alt["estres_alt_pct_raw"].clip(lower=0, upper=100)

alt_stats = pd.DataFrame(
    {
        "metric": [
            "rows_total",
            "rows_non_null_raw",
            "min_raw",
            "max_raw",
            "rows_above_100_raw",
            "rows_non_null_clipped",
            "min_clipped",
            "max_clipped",
        ],
        "value": [
            len(rha_alt),
            rha_alt["estres_alt_pct_raw"].notna().sum(),
            rha_alt["estres_alt_pct_raw"].min(skipna=True),
            rha_alt["estres_alt_pct_raw"].max(skipna=True),
            (rha_alt["estres_alt_pct_raw"] > 100).sum(skipna=True),
            rha_alt["estres_alt_pct_clipped"].notna().sum(),
            rha_alt["estres_alt_pct_clipped"].min(skipna=True),
            rha_alt["estres_alt_pct_clipped"].max(skipna=True),
        ],
    }
)

print("Current proxy range check")
display(current_stats)
print("\nAlternative proxy range check")
display(alt_stats)
print("\nNon-consumptive REPDA columns detected")
print({"subterraneo": sub_non_cons_cols, "superficial": sup_non_cons_cols})
````

## QC. Compare against official 6.4.2 and calibrate

Compare national aggregate from proxy components against official SDG 6.4.2 values and derive a calibration factor.


````python
# Official national series from INEGI/ODS file
official_file = "../datos/6.4.2/conjunto_de_datos/6.4.2_sh_es.csv"
official_sh = pd.read_csv(official_file, usecols=[0, 1])
official_sh.columns = ["anio", "stress_official_pct"]
official_sh["anio"] = pd.to_numeric(official_sh["anio"], errors="coerce")
official_sh["stress_official_pct"] = (
    official_sh["stress_official_pct"].astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(r"[^\d.\-]", "", regex=True)
)
official_sh["stress_official_pct"] = pd.to_numeric(official_sh["stress_official_pct"], errors="coerce")
official_sh = official_sh.dropna().copy()
official_sh["anio"] = official_sh["anio"].astype(int)

# National aggregate from alternative proxy components (ratio of sums)
national_proxy = (
    rha_alt.groupby("anio", as_index=False)[
        ["TFWW_consuntivo_hm3", "TRWR_alt_hm3", "EFR_alt_hm3", "available_alt_hm3"]
    ]
    .sum()
)
national_proxy["stress_proxy_pct"] = np.where(
    national_proxy["available_alt_hm3"] > 0,
    national_proxy["TFWW_consuntivo_hm3"] / national_proxy["available_alt_hm3"] * 100,
    np.nan,
)

comparison = national_proxy.merge(official_sh, on="anio", how="inner")
comparison = comparison.sort_values("anio").reset_index(drop=True)
comparison["abs_error_pp"] = (comparison["stress_proxy_pct"] - comparison["stress_official_pct"]).abs()
comparison["ratio_official_to_proxy"] = np.where(
    comparison["stress_proxy_pct"] > 0,
    comparison["stress_official_pct"] / comparison["stress_proxy_pct"],
    np.nan,
)

# Single calibration factor over overlap years
calibration_factor = comparison["ratio_official_to_proxy"].median(skipna=True)

# Apply correction to RHA disaggregated series (NO clipping; preserve >100 cases)
rha_alt["estres_alt_pct_corrected"] = rha_alt["estres_alt_pct_raw"] * calibration_factor

calibration_summary = pd.DataFrame(
    {
        "metric": [
            "overlap_years",
            "calibration_factor_median",
            "proxy_min_before",
            "proxy_max_before",
            "proxy_min_after",
            "proxy_max_after",
            "rows_above_100_after",
        ],
        "value": [
            ", ".join(map(str, comparison["anio"].tolist())),
            calibration_factor,
            rha_alt["estres_alt_pct_raw"].min(skipna=True),
            rha_alt["estres_alt_pct_raw"].max(skipna=True),
            rha_alt["estres_alt_pct_corrected"].min(skipna=True),
            rha_alt["estres_alt_pct_corrected"].max(skipna=True),
            (rha_alt["estres_alt_pct_corrected"] > 100).sum(skipna=True),
        ],
    }
)

print("National comparison: proxy vs official")
display(comparison[["anio", "stress_proxy_pct", "stress_official_pct", "abs_error_pp", "ratio_official_to_proxy"]])
print("\nCalibration summary")
display(calibration_summary)
````
